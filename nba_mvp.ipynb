```python
from nba_api.stats.endpoints import playercareerstats as pcs
import pandas as pd

# Player IDs
jokic_id = '203999'  
gilgeous_alexander_id = '1628983'  

# Fetch career stats for both players

jokic_stats = pcs.PlayerCareerStats(player_id=jokic_id)
gilgeous_stats = pcs.PlayerCareerStats(player_id=gilgeous_alexander_id)

# Get career totals data as DataFrames
jokic_career_totals_df = jokic_stats.get_data_frames()[1]  # Remember Index 1 is usually the career totals DataFrame
gilgeous_career_totals_df = gilgeous_stats.get_data_frames()[1]

jokic_car_totals = jokic_stats.get_data_frames()[1]

# Combine the data into a single DataFrame for comparison
combined_df = pd.concat([jokic_career_totals_df, gilgeous_career_totals_df])
combined_df.index = ['Nikola Jokic', 'Shai Gilgeous-Alexander']

# Print the combined DataFrame
combined_df.head().T
```

```python
#Now we want to analyze the stats for the 2023-2024 season
from nba_api.stats.endpoints import playergamelog
import pandas as pd

# Player IDs
jokic_id = '203999'  # Nikola Jokic
gilgeous_alexander_id = '1628983'  # Shai Gilgeous-Alexander

# Season specification
season = '2023-24'

# Fetch game logs for the 2023-2024 season for both players
jokic_logs = playergamelog.PlayerGameLog(player_id=jokic_id, season=season)
gilgeous_logs = playergamelog.PlayerGameLog(player_id=gilgeous_alexander_id, season=season)

# Get the game logs data as DataFrames
jokic_games_df = jokic_logs.get_data_frames()[0]  # The first DataFrame contains the game logs
gilgeous_games_df = gilgeous_logs.get_data_frames()[0]

# Calculate season averages for each player
jokic_season_avg = jokic_games_df.mean(numeric_only=True)
gilgeous_season_avg = gilgeous_games_df.mean(numeric_only=True)

# Combine the averages into a single DataFrame for comparison
avg_stats_df = pd.DataFrame({
    'Nikola Jokic': jokic_season_avg,
    'Shai Gilgeous-Alexander': gilgeous_season_avg
})

# Print the combined DataFrame
avg_stats_df.T
#gilgeous_games_df.info()
```

```python
#Now we're analyzing OKC VS DEN stats
from nba_api.stats.endpoints import teamyearbyyearstats
import pandas as pd
import matplotlib.pyplot as plt

# Team IDs
okc_id = '1610612760'  # Oklahoma City Thunder
den_id = '1610612743'  # Denver Nuggets

# Fetch year-by-year stats for both teams
okc_stats = teamyearbyyearstats.TeamYearByYearStats(team_id=okc_id).get_data_frames()[0]
den_stats = teamyearbyyearstats.TeamYearByYearStats(team_id=den_id).get_data_frames()[0]

# Filter for the 2023-2024 season
okc_2023_2024 = okc_stats[okc_stats['YEAR'] == '2023-24']
den_2023_2024 = den_stats[den_stats['YEAR'] == '2023-24']

# Combine the data into a single DataFrame for comparison
combined_df = pd.concat([okc_2023_2024, den_2023_2024])
combined_df.index = ['OKC', 'DEN']


combined_df.T
```
